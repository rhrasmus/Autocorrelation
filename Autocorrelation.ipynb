```python
from random import *
from matplotlib import pyplot
from statsmodels.graphics.tsaplots import plot_acf
import pandas as pd
from arch import arch_model
```

```python
seed(20000)
data = [gauss(0,i*0.01) for i in range(0,100)]
squared_data = [x**2 for x in data]
#pd.DataFrame(squared_data).tail()
plot_acf(pd.array(squared_data),lags=99)
pyplot.show()
pyplot.plot(squared_data)
pyplot.show()
```

```python
tests = 10
model = arch_model(data, mean = 'Zero', vol = 'ARCH', p = 15)
model_fit = model.fit()
y_hat = model_fit.forecast(horizon=tests)
pyplot.plot(y_hat.variance.values[-1, :])
pyplot.show()
```
